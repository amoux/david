```python
import os
import re
import pandas as pd

from vuepoint.comment_miner import download_comments
from vuepoint.youtube_api import youtube_search_video
```

```python
download_comments(youtube_id='hixW12aWRRU', file_name='test')
```

```python
def make_query_dir(search_query):
    query_dir_name = re.sub(' ', '', search_query)
    query_dir_path = 'downloads/' + query_dir_name
    return query_dir_path


def download_all_videoids(ids_list, search_query):
    query_dir_path = make_query_dir(search_query)
    print(f'\nSaving all downloads in directory path: {query_dir_path}')
    print(f'downloading comments from: {len(ids_list)} videos')
    for ids in ids_list:
        download_comments(youtube_id=ids,
                          file_name=str(ids),
                          limit=10,
                          dir_name=str(query_dir_path))


def saveQueryToCSV(df, search_query):
    dir_name = re.sub(' ', '', search_query)
    dir_path = make_query_dir(dir_name)
    save_path = os.path.join(dir_path, dir_name)
    save_file = save_path + '.csv'
    df.to_csv(save_file)
    print(f'"{search_query}" saved in path: {save_file}')
```

```python
SEARCH_QUERY = 'Huawei'
```

```python
# returns multiple channel results from a search query
video_list = youtube_search_video(SEARCH_QUERY, max_results=10)

# pass the search query to pandas & get only videos with comment len > 2000
df = pd.DataFrame.from_dict(video_list, orient='columns')
#df['commentCount'] = df['commentCount'].astype(int)
#df = df[df.commentCount > 2000]
df
```

```python
# get the video id list from pandas
video_ids_list = list(df['vidId'])

# download comments for each video id in the list
download_all_videoids(video_ids_list, SEARCH_QUERY)

# save search query to a csv file
saveQueryToCSV(df, SEARCH_QUERY)
```
